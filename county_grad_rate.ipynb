```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
grad = pd.read_csv("../data/grad_rate.csv")
district = pd.read_csv("../data/district_county.csv")
```

```python
district
```

```python
grad = grad[["Graduating Class", "District", "Graduation Rate"]]
```

```python
grad.rename({'Graduating Class': 'year'}, axis=1, inplace=True)
```

```python
district = district[["County Name", "District"]]
```

```python
grad_county = grad.merge(district, on='District').groupby(['County Name', 'year'])['Graduation Rate'].mean().reset_index()
```
